```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Create a tiny sample dataset
data = {
    'title': ['Movie A', 'Movie B', 'Movie C'],
    'tags': ['action thriller', 'action romance', 'romance comedy']
}
df = pd.DataFrame(data)

print("--- OUR MOVIES ---")
display(df) # 'display' makes tables look beautiful in Jupyter

# 2. Turn the text into mathematical vectors
cv = CountVectorizer()
vectors = cv.fit_transform(df['tags']).toarray()

print("\n--- THE VOCABULARY ---")
print("Words the AI learned:", cv.get_feature_names_out())

print("\n--- THE VECTORS ---")
print(vectors)

# 3. Calculate the Cosine Similarity (The angles between the vectors)
similarity = cosine_similarity(vectors)

print("\n--- SIMILARITY MATRIX ---")
print(similarity)
```
